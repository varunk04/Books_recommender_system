### Lets build a poularity based recommender system:

* We will count the total number of ratings for each book.

* Then we will calculate the mean of the ratings.

* we will only consider books having atleat 250 ratings to it.

* Finally recommend top 50 books according to above criteria.

```python
# To do this we will use books and ratings datasets:

ratings_with_name = ratings.merge(books, on ='ISBN')
```

```python
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating':'num_ratings'}, inplace = True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns = {'Book-Rating':'avg_ratings'}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df
```

These are top 50 books recommendation to everyone, it is common through out every users.

```python
# Lets merge popular_df with books to get the details about books.

popular_df = popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popular_df.head()
```

## Collaborative Filtering Based Recommender System:

* We wil try to filter out the users who has voted for more than 250 times, since they have experice in reading books.

* Also we will consider the books which has more than 50 ratings to it, inorder to consider only popular books.


```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200

# This will give all the users who has voted more than 200 times
top_users = x[x].index
```

```python
len(top_users)
```

There are 811 users

```python
# now lets find the books that these 811 users has rated.

filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(top_users)]
```

```python
filtered_ratings
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >=50

famous_books = y[y].index
```

```python
len(famous_books)
```

There are 706 books which has more than 50 ratings to it.

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

My final_df has details about books whihc has been rated more than 50 times by the users who has rated the books more than 250 times.

```python
# Lets pivot the table:

pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')

pt
```

Lets fill missing values with 0.

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
similarity_score
```

```python
'''
Lets build a function which will taek book_name as the parameters 
and return 5 books that has similarity scores with that particluar based on the above array.

'''
def recommend(book_name):
    #fetching_index
    index = np.where(pt.index == book_name)[0][0]
    # Getching 5 books which has highest similarity scores.
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x : x[1], reverse = True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])   
```

```python
recommend('1984')
```

### Recommen function recommends the 5 films that are very similiar to the movie which is being passed to the function, this is a simple way of building a recommender systems.
